```python
csv_file.groupby(["Region"]).sum().sort_values("Total population",ascending=True).plot(kind="bar",figsize=(15,10))
```

```python
#totaldoses for astra and moderna doses without population
all_doses = csv_file.groupby("Region").sum()
```

```python
all_doses
```

```python
all_doses.insert(4,"all region",["North Central","North East","North West","South East","South South","South West"])
```

```python
all_doses
```

```python
all_doses.plot(kind="line",x="all region",y="Total Doses Administered (Astra)",marker="*",figsize=(15,10))
```

```python
all_doses.sort_values("Total Doses Administered (Moderna)",ascending=True).plot(kind="barh",x="all region",y="Total Doses Administered (Moderna)",color="orange")
```

```python
all_doses.sort_values("Total Doses Administered (Moderna)",ascending=True).plot(kind="area",x="all region",y="Total Doses Administered (Moderna)",color="green")
```

```python
all_doses.sort_values("Total Doses Administered (Moderna)",ascending=True).plot(kind="barh",x="all region",color=["orange","brown","blue","tomato"])
```
